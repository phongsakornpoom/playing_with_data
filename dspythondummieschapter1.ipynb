```python
with open('Colors.txt','r') as loop:
    for loopop in loop:
        print('containing: ',loopop)#object in for loop no need to use read()
    
```

```python
with open('Colors.txt','r') as indexloop:
    for j,index in enumerate(indexloop):#enumerate is use for reading line we want
        #j or any variable is the index that start without number is 0
        if j%2==0:#condition which line to read
            print('Reading line',str(j),'content',index)#printing output
            
```
